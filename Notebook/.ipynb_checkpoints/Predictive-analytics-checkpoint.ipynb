### Exploratory Data Analysis (EDA)

```python
df = pd.read_csv("Cleaned_SampleDataset.csv")
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['Sales'].describe()
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Month_Name'] = df['Date'].dt.strftime('%b')
```

```python
df[['Date','Year','Month','Month_Name']].head()
```

#### Overall Sales Trend

```python
plt.figure(figsize=(12,6))
plt.plot(df['Date'], df['Sales'])
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
monthly_sales = df.groupby(
    pd.Grouper(key='Date', freq='ME')
)['Sales'].sum().reset_index()

monthly_sales.head()
```

```python
plt.figure(figsize=(12,6))

plt.plot(monthly_sales['Date'],
         monthly_sales['Sales'],
         marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Sales")

plt.grid(True)

plt.show()
```

#### Monthly Sales Analysis

```python
monthly_sales = df.groupby('Month_Name')['Sales'].sum()
monthly_sales
```

```python
monthly_sales.plot(
    kind='bar',
    figsize=(10,5)
)

plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

#### Yearly Sales Analysis

```python
yearly_sales = df.groupby('Year')['Sales'].sum()

yearly_sales
```

```python
yearly_sales.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Yearly Sales")
plt.ylabel("Sales")
plt.show()
```

#### Category Analysis

```python
category_sales = df.groupby('Category')['Sales'].sum()

category_sales
```

```python
category_sales.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Sales by Category")
plt.ylabel("Sales")
plt.show()
```

#### Region Analysis

```python
region_sales = df.groupby('Region')['Sales'].sum()

region_sales
```

```python
region_sales.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Sales by Region")
plt.ylabel("Sales")
plt.show()
```
