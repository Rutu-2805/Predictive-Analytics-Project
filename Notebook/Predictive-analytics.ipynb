## Prediction model

```python
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv( "Output_Monthly_sales.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date']= pd.to_datetime(df['Date'])
```

```python
df['Month_Number'] = range(len(df))
```

```python
df.head()
```

```python
x = df[['Month_Number']]      # Input
```

```python
y = df['Sales']             # output
```

```python
print(x.head())
print(y.head())
```

```python
model = LinearRegression()   # to initialize a "Linear Regression" object and stores it inside a variable named model
```

```python
model.fit(x,y)
```

```python
df['Predicted_Sales'] = model.predict(x)
```

```python
df.head()
```

```python
plt.figure(figsize=(12,6))

plt.plot(
    df['Date'],
    df['Sales'],
    label='Actual Sales'
)

plt.plot(
    df['Date'],
    df['Predicted_Sales'],
    label='Predicted Sales'
)

plt.title("Actual vs Predicted Sales")
plt.xlabel("Date")
plt.ylabel("Sales")

plt.legend()

plt.show()
```

```python
future_months = pd.DataFrame({
    'Month_Number': range(
        len(df),
        len(df)+12
    )
})                                               #Future months numbers
```

```python
future_months          
```

```python
future_predictions = model.predict(
    future_months
)                                            #Future predictions
```

```python
print(future_predictions)
```

```python
last_date = df['Date'].max()

print(last_date)                   #Future dates
```

```python
future_dates = pd.date_range(
    start=last_date,
    periods=13,
    freq='ME'
)[1:]
```

```python
future_dates
```

```python
forecast_df
```

```python
plt.figure(figsize=(12,6))

plt.plot(
    df['Date'],
    df['Sales'],
    label='Historical Sales'
)

plt.plot(
    forecast_df['Date'],
    forecast_df['Forecasted_Sales'],
    label='Future Forecast'
)

plt.title("Sales Forecast for Next 12 Months")
plt.xlabel("Date")
plt.ylabel("Sales")

plt.legend()

plt.show()
```

## Model Evaluation and Visualization of data

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score 
```

```python
x = df[['Month_Number']]
y = df['Sales']
```

```python
split_index = int(len(df) * 0.8)

x_train = x[:split_index]
x_test = x[split_index:]

y_train = y[:split_index]
y_test = y[split_index:]
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
```

```python
print(pred[:5])
```

```python
mae = mean_absolute_error(
    y_test,
    pred
)

print("MAE:", mae)
```

```python
mse = mean_squared_error(
    y_test,
    pred
)

print("MSE:", mse)
```

```python
rmse = mse ** 0.5

print("RMSE:", rmse)
```

```python
r2 = r2_score(
    y_test,
    pred
)

print("R² Score:", r2)
```

```python
plt.scatter(x_test, y_test)
plt.scatter(x_test, pred)
plt.show()
```

```python
results = pd.DataFrame({
    'Actual': y_test,
    'Predicted': pred
}).sort_index()

plt.figure(figsize=(12,6))
plt.plot(results['Actual'].values, label='Actual')
plt.plot(results['Predicted'].values, label='Predicted')
plt.legend()
plt.show()
```

```python
results.to_csv(
    "Model_evaluation.csv",
    index=False
)
```
